# Plotting with gwrefpy ðŸ“Š

This quick example demonstrates how to use the plotting capabilities of `gwrefpy`.

We will go step by step:
1. Import the example data
2. Plot the data with a few different options 

First we import the package and initialize a model and load the example data.

```python
import gwrefpy as gr

model = gr.Model("small_example.gwref")
```

This example data contains 4 wells and 2 fits. We can see this by printing the model object.

```python
model
```

## Plotting well timeseries data ðŸ“ˆ

Now we can plot the data. We start by plotting all of the wells in the same figure.

```python
model.plot_wells(plot_style="fancy")
```

We can see that the labels for the two reference wells are overlapping. We can fix this by simply adding an offset to the text labels. Here we add a small offset to the two reference wells.

```python
offsets = {
    "46LOGW":-.2,
    "45LOGW":.2
}
model.plot_wells(plot_style="fancy", offset_text=offsets)
```

We can also plot wells by passing them as a list. Here we plot only the two reference wells.

```python
model.plot_wells(wells=model.ref_wells, plot_style="fancy")
```

## Plotting fitted data ðŸ“‰

Now we can plot the fits. We can plot all of the fits in the model by simply calling the `plot_fits` method. The filled areas represent the prediction interval of each fit and the red circles mark outliers.

```python
model.plot_fits(plot_style="fancy", offset_text=offsets)
```

We can also plot them individually by passing only a sinlge fit to the function or by passing a list of fits and seeting the `plot_Seperately` argument to `True`. We can also show the initiation period by setting the `show_initiation` argument to `True`.

```python
model.plot_fits(plot_style="fancy", plot_separately=True, show_initiation_period=True)
```

You can also change the color theme of the plots by changing the default `'color'` to `'monochrome`, and change the style from the default `'fancy'` to `'scientific'`. Not that changing the `color_theme` will reset and reassign all plot attributes, include the user specified.

```python
model.plot_fits(color_style='monochrome', plot_style='scientific', plot_separately=True, show_initiation_period=True)
```
